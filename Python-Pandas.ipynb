# Dia 1

## Importando os dados

```python
# Importando a biblioteca
import pandas as pd
```

## Lendo os arquivos csv

```python
# CSV
anos = [20101, 20111, 20121, 20131, 20141, 20151, 20161, 20171, 20181, 20191, 20201, 20102, 20112, 20122, 20132, 20142, 20152, 20162, 20172, 20182, 20192]

# Lista para armazenar os DataFrames
dfs = []

for ano in anos:
    link_csv = f'https://raw.githubusercontent.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/main/Dia_1-Importando_dados/Datasets/dados_emprestimos/emprestimos-{ano}.csv'
    df_csv = pd.read_csv(link_csv)
    dfs.append(df_csv)
```

## Concatenando os DataFrames

```python
df_csv_concat = pd.concat(dfs, ignore_index=True)
df_csv_concat
```

## Verificando os valores duplicados 

```python
df_csv_concat.value_counts()
```

## Excluindo valores duplicados dos arquivos csv

```python
df_csv_concat = df_csv_concat.drop_duplicates()
df_csv_concat.value_counts()
```

```python
df_csv_concat.head()
```

## Importando mais dados

```python
dados_exemplares = pd.read_parquet('https://raw.githubusercontent.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet')
dados_exemplares
```

## Excluindo dados duplicados do arquivo parquet

```python
dados_exemplares = dados_exemplares.drop_duplicates()
dados_exemplares
```

## Unindo os DataFrames

```python
emprestimos_completos = df_csv_concat.merge(dados_exemplares)
emprestimos_completos
```

# Dia 2

## Excluindo a coluna "registro_sistema"

```python
emprestimos_completos.drop('registro_sistema', axis=1, inplace=True)
emprestimos_completos
```

## Convertendo a coluna "matricula_ou_siape" para String

```python
emprestimos_completos['matricula_ou_siape'] = emprestimos_completos['matricula_ou_siape'].astype('string')
emprestimos_completos.head()
```

## Atribuindo a coluna CDU_geral

```python
CDU_lista = []
for CDU in emprestimos_completos['localizacao']:
    if (CDU < 100):
        CDU_lista.append('Generalidades. Ciência e conhecimento')
    elif (CDU < 200):
        CDU_lista.append('Filosofia e psicologia')
    elif (CDU < 300):
        CDU_lista.append('Religião')
    elif (CDU < 400):
        CDU_lista.append('Ciências sociais')
    elif (CDU < 500):
        CDU_lista.append('Classe vaga. Provisoriamente não ocupada')
    elif (CDU < 600):
        CDU_lista.append('Matemática e ciências naturais')
    elif (CDU < 700):
        CDU_lista.append('Ciências aplicadas')
    elif (CDU < 800):
        CDU_lista.append('Belas artes')
    elif (CDU < 900):
        CDU_lista.append('Linguagem. Língua. Linguística')
    else:
        CDU_lista.append('Geografia. Biografia. História')

emprestimos_completos['CDU_geral'] = CDU_lista
```
