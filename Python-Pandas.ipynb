# Dia 1

## Importando os dados

```python
# Importando a biblioteca
import pandas as pd
```

## Lendo os arquivos csv

```python
# CSV
anos = [20101, 20111, 20121, 20131, 20141, 20151, 20161, 20171, 20181, 20191, 20201, 20102, 20112, 20122, 20132, 20142, 20152, 20162, 20172, 20182, 20192]

# Lista para armazenar os DataFrames
dfs = []

for ano in anos:
    link_csv = f'https://raw.githubusercontent.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/main/Dia_1-Importando_dados/Datasets/dados_emprestimos/emprestimos-{ano}.csv'
    df_csv = pd.read_csv(link_csv)
    dfs.append(df_csv)
```

## Concatenando os DataFrames

```python
df_csv_concat = pd.concat(dfs, ignore_index=True)
df_csv_concat
```

## Verificando os valores duplicados 

```python
df_csv_concat.value_counts()
```

## Excluindo valores duplicados dos arquivos csv

```python
df_csv_concat = df_csv_concat.drop_duplicates()
df_csv_concat.value_counts()
```

```python
df_csv_concat.head()
```

## Importando mais dados

```python
dados_exemplares = pd.read_parquet('https://raw.githubusercontent.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet')
dados_exemplares
```

## Excluindo dados duplicados do arquivo parquet

```python
dados_exemplares = dados_exemplares.drop_duplicates()
dados_exemplares
```

## Unindo os DataFrames

```python
emprestimos_completos = df_csv_concat.merge(dados_exemplares)
emprestimos_completos
```

# Dia 2

## Excluindo a coluna "registro_sistema"

```python
emprestimos_completos.drop('registro_sistema', axis=1, inplace=True)
emprestimos_completos
```

## Convertendo a coluna "matricula_ou_siape" para String

```python
emprestimos_completos['matricula_ou_siape'] = emprestimos_completos['matricula_ou_siape'].astype('string')
emprestimos_completos.head()
```

## Atribuindo a coluna CDU_geral

```python
CDU_lista = []
for CDU in emprestimos_completos['localizacao']:
    if (CDU < 100):
        CDU_lista.append('Generalidades. Ciência e conhecimento')
    elif (CDU < 200):
        CDU_lista.append('Filosofia e psicologia')
    elif (CDU < 300):
        CDU_lista.append('Religião')
    elif (CDU < 400):
        CDU_lista.append('Ciências sociais')
    elif (CDU < 500):
        CDU_lista.append('Classe vaga. Provisoriamente não ocupada')
    elif (CDU < 600):
        CDU_lista.append('Matemática e ciências naturais')
    elif (CDU < 700):
        CDU_lista.append('Ciências aplicadas')
    elif (CDU < 800):
        CDU_lista.append('Belas artes')
    elif (CDU < 900):
        CDU_lista.append('Linguagem. Língua. Linguística')
    else:
        CDU_lista.append('Geografia. Biografia. História')

emprestimos_completos['CDU_geral'] = CDU_lista
```

## Quantidade total de exemplares emprestados por ano

```python
emprestimo_data = pd.DataFrame(emprestimos_completos['data_emprestimo'].value_counts()).reset_index()
emprestimo_data.columns = ['data', 'quantidade']
emprestimo_data['data'] = pd.to_datetime(emprestimo_data['data'])
emprestimo_data
```

```python
emprestimo_por_ano = emprestimo_data.groupby(by=emprestimo_data.data.dt.year)['quantidade'].sum()
emprestimo_por_ano.index.name = 'ano'
emprestimo_por_ano = pd.DataFrame(emprestimo_por_ano)
emprestimo_por_ano
```

## Plotando gráfico por ano

```python
# Importando as bibliotecas
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
sns.set_theme(context='notebook', 
            style='darkgrid', 
            palette='deep', 
            font_scale=1.3, 
            rc={"figure.figsize":(15,8)})
```

```python
ax = sns.lineplot(data=emprestimo_por_ano,x='ano',y='quantidade')
ax.set(xlabel=None,ylabel=None)
ax.tick_params(axis='x', rotation=30)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))   

ax.set_title('Quantidade de exemplares emprestados do SISBI por ano'+'\n',size=20,loc='left',weight='bold')

plt.show()
```

Pode-ser perceber que desde 2013 os números de exemplares emprestados diminuindo, com uma queda bem grande em 2019.

## Emprestimos por mês

```python
emprestimos_por_mes = emprestimo_data.groupby(by=emprestimo_data.data.dt.month)['quantidade'].sum()
emprestimos_por_mes.index.name = 'mes'
emprestimos_por_mes = pd.DataFrame(emprestimos_por_mes)
emprestimos_por_mes
```

### Trocando os números por nomes dos meses

```python
dicionario_meses = {1:'Jan', 2:'Fev', 3:'Mar', 4:'Abr',
                    5:'Mai', 6:'Jun', 7:'Jul', 8:'Ago',
                    9:'Set',10:'Out',11:'Nov', 12:'Dez'}
```

```python
emprestimos_por_mes.index = emprestimos_por_mes.index.map(dicionario_meses)
```

```python
emprestimos_por_mes
```

## Plotando gráfico por mês

```python
ax = sns.lineplot(data=emprestimos_por_mes, x='mes', y='quantidade')
ax.set(xlabel=None, ylabel=None)
ax.tick_params(axis='x', rotation=30)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))

ax.set_title('Quantidade de exemplares emprestados por mês'+'\n', size=20, loc='left', weight='bold')
ax.text(s='Período de 2010 a 2020', x=0.5, y=265000, ha='left', color='grey')

plt.show()
```

Pode-se visualizar que os meses com maiores números de exemplares emprestados foram em março e agosto.

Por ser uma biblioteca universitária meses de férias como janeiro, julho e dezembro são os menores números.

O que não é nada de novo, entretanto caso todos os colaboradores saírem de férias durante esse período não conseguiríamos manter todo o funcionamento da biblioteca.

Principalmente com atividades de inventário que podem ocorrer anualmente ou de dois em dois anos e o melhor período seria de dezembro a janeiro.

Então meses como maio, junho, outubro e novembro podem ser uma opção a ser considerada pela direção da biblioteca.

Nota-se também a grande queda no mês de junho. Cabe aqui um alerta para se investigar mais, o que está ocorrendo e desenvolver estratégias e ações de marketing da biblioteca para que se aumente o número de exemplares emprestados.

Como por exemplo: uma ação para que sejam realizados empréstimos de livros para a leitura nas férias de julho.

## Quantidade de empréstimos por hora ao longo do dia

```python
emprestimo_data.data
```

```python
emprestimos_por_hora = emprestimo_data.groupby(by=emprestimo_data.data.dt.hour)['quantidade'].sum()
emprestimos_por_hora.index.name = 'horas'
emprestimos_por_hora = pd.DataFrame(emprestimos_por_hora).reset_index()
emprestimos_por_hora
```

```python
emprestimos_por_hora = emprestimos_por_hora.sort_values(ascending=True,by='quantidade')
```

```python
ax = sns.barplot(data=emprestimos_por_hora,y='quantidade',x='horas',
                palette='Blues',hue='quantidade',dodge=False)                  #Ordenar pela quantidade de exemplares emprestados
plt.legend([],[], frameon=False)                                                #Excluir a legenda do gráfico

ax.set(xlabel='Horário',ylabel=None)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))           
ax.set_title("Quantidade de exemplares emprestados do SISBI por faixa horária"+"\n",size=20,loc='left',weight='bold')
ax.text(s='Período de 2010 a 2020',x=-0.5,y=225000,fontsize=18, ha='left',color='gray')
ax=ax
```

Através do gráfico é possível identificar que os períodos das 10 h e 16 h são os de maior pico no atendimento de empréstimos.

Períodos do começo da manhã (entre 7h e 9) e da noite (entre 19h e 22) podem ser os melhores horários para que se realizem as demais atividades além do atendimento.

## Criando a função para gerar as tabelas

```python
def tabela_frequencia(variavel):
    dataframe = pd.DataFrame(emprestimos_completos[variavel].value_counts())
    dataframe.columns = ['quantidade']
    dataframe['percentual'] = round((dataframe.quantidade / dataframe.quantidade.sum())*100,1)

    return dataframe
```

## Quantos foram pelo tipo de vínculo dos usuários ?

```python
tabela_frequencia('tipo_vinculo_usuario')
```

O maior público da biblioteca são os alunos de graduação seguidos pelos de pós-graduação.

## Quantos foram pelas coleções ?

```python
tabela_frequencia('colecao')
```

A principal coleção é a circulante com 99% de exemplares emprestados.

Algo que poderiamos entender é por que as outras coleções não estão sendo consultadas
poderiamos fazer isso através de uma pesquisa para saber, se os documentos disponíveis
não são pegos fisicamente pois são utilizados de forma online.

## Quantos foram por biblioteca ?

```python
tabela_frequencia('biblioteca')
```

A biblioteca centrar tem maior predominancia, o que já era de se esperar
devido ao seu tamanho.

## Quantos foram pela CDU geral
