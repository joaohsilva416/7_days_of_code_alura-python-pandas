## Atribuindo a coluna CDU_geral

```python
CDU_lista = []
for CDU in emprestimos_completos['localizacao']:
    if (CDU < 100):
        CDU_lista.append('Generalidades. Ciência e conhecimento')
    elif (CDU < 200):
        CDU_lista.append('Filosofia e psicologia')
    elif (CDU < 300):
        CDU_lista.append('Religião')
    elif (CDU < 400):
        CDU_lista.append('Ciências sociais')
    elif (CDU < 500):
        CDU_lista.append('Classe vaga. Provisoriamente não ocupada')
    elif (CDU < 600):
        CDU_lista.append('Matemática e ciências naturais')
    elif (CDU < 700):
        CDU_lista.append('Ciências aplicadas')
    elif (CDU < 800):
        CDU_lista.append('Belas artes')
    elif (CDU < 900):
        CDU_lista.append('Linguagem. Língua. Linguística')
    else:
        CDU_lista.append('Geografia. Biografia. História')

emprestimos_completos['CDU_geral'] = CDU_lista
```

## Função para gerar o gráfico de boxplot

```python
def gerar_box_plot(dataset, x, y, titulo, subtitulo):
    sns.set_theme(style='white', palette='tab10', font_scale=1.2)
    plt.figure(figsize=(16,10))
    
    ax = sns.boxplot(data=dataset, x=x, y=y)
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))

    plt.ylim(0,max(dataset[y])*1.1)
    plt.xlabel(None)
    plt.ylabel(None)

    ax.set_title(titulo+'\n', fontsize=20, loc='left', weight='bold')
    ax.text(s=subtitulo,x=-0.5, y=max(dataset[y])*1.11, fontsize=18, ha='left', color='grey')
```

## Plotando gráfico dos Alunos de Graduação

```python
gerar_box_plot(alunos_graduacao_acervo_circulante, 'ano', 'quantidade', 'Distribuição dos empréstimos mensais', 'Realizados pelos alunos de graduação no acervo circulante')
```

## Criando tabela em HTML

### Estruturando formatação da tabela em CSS

```python
th_props = [
    ('font-size', '1.4rem'),
    ('text-align', 'center'),
    ('font-weight', 'bold'),
    ('color', 'whitesmoke'),
    ('background-color', '#001692'),
    ('border-radius', '0.25rem'),
    ('box-shadow','0 0 1rem gray')
]

td_props = [
    ('font-size', '1rem'),
    ('padding','0.5rem'),
    ('text-align', 'left'),
    ('font-weight', 'bold'),
    ('border-bottom','0.1rem solid lightgray')
]

styles = [
    dict(selector="th", props=th_props),
    dict(selector="td", props=td_props)
]
```

## Gerando tabela em HTML

```python
percentual.style.text_gradient(cmap='RdYlGn',low=1, axis=1,vmax=0.1,vmin=0)\
                                .format('{:.2f} %',subset=['2018','2019','2022'])\
                                .set_table_styles(styles)\
                                .to_html('teste.html',doctype_html =True,
                                        table_attributes='ALIGN=LEFT WIDTH=50% CELLSPACING = 5')
```
